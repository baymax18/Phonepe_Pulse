```python
## Aggregate Transcation

path = "F:/phonepay/pulse/data/aggregated/transaction/country/india/state"
Agg_state_list=os.listdir(path)
Agg_state_list
column = {'state':[], 'year':[],'quarter':[],'transaction_type':[], 'transaction_count':[], 'transaction_amount':[]}
for state in Agg_state_list:
    p_i = path+'/'+ state +'/'
    Agg_yr = os.listdir(p_i)
    for year in Agg_yr:
        p_j=p_i+year+"/"
        Agg_yr_list=os.listdir(p_j)
        for file in Agg_yr_list:
            p_k=p_j+file
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              column['transaction_type'].append(str(Name))
              column['transaction_count'].append(count)
              column['transaction_amount'].append(amount)
              column['state'].append(str(state))
              column['year'].append(int(year))
              column['quarter'].append(int(file.strip('.json')))

Agg_Trans=pd.DataFrame(column)
name_alter(Agg_Trans)
Agg_Trans
```

```python
## Aggregate User

path = "F:/phonepay/pulse/data/aggregated/user/country/india/state"
Agg_state_list=os.listdir(path)
Agg_state_list
column2 = {'state':[], 'year':[],'quarter':[], 'brand':[],'devicecount':[],'percentage':[]}
for state in Agg_state_list:
    p_i = path+'/'+ state +'/'
    Agg_yr = os.listdir(p_i)
    for year in Agg_yr:
        p_j=p_i+year+"/"
        Agg_yr_list=os.listdir(p_j)
        for file in Agg_yr_list:
            p_k=p_j+file
            Data=open(p_k,'r')
            D=json.load(Data)
            try:
                for z in D['data']['usersByDevice']:
                    brand = z['brand']
                    devicecount = z['count']
                    percentage = z['percentage']
                    column2['brand'].append(brand)
                    column2['devicecount'].append(devicecount)               
                    column2['percentage'].append(percentage)
                    column2['state'].append(state)
                    column2['year'].append(int(year))
                    column2['quarter'].append(int(file.strip('.json')))
            except:
                pass

Agg_user=pd.DataFrame(column2)
name_alter(Agg_user)
Agg_user
```

```python
### Agg Insurance

path = "F:/phonepay/pulse/data/aggregated/insurance/country/india/state"
Agg_state_list=os.listdir(path)
Agg_state_list
column3 = {'state':[], 'year':[],'quarter':[], 'count':[],'amount':[]}
for state in Agg_state_list:
    p_i = path+'/'+ state +'/'
    Agg_yr = os.listdir(p_i)
    for year in Agg_yr:
        p_j=p_i+year+"/"
        Agg_yr_list=os.listdir(p_j)
        for file in Agg_yr_list:
            p_k=p_j+file
            Data=open(p_k,'r')
            D=json.load(Data)
            for i in D['data']['transactionData']:
                count = (i['paymentInstruments'][0]['count'])
                column3['count'].append(count)
                amount = (i['paymentInstruments'][0]['amount'])
                column3['amount'].append(amount)
                column3['state'].append(state)
                column3['year'].append(int(year))
                column3['quarter'].append(int(file.strip('.json')))
Agg_Insurance=pd.DataFrame(column3)
name_alter(Agg_Insurance)
Agg_Insurance
```

```python
### map Insurance

path = "F:/phonepay/pulse/data/map/insurance/country/india/state"
map_state_list=os.listdir(path)

column4 = {'state':[], 'year':[],'quarter':[], 'lat':[],'lng':[],'metric':[],'label':[]}
for state in map_state_list:
    p_i = path+'/'+ state +'/'
    map_yr = os.listdir(p_i)
    for year in map_yr:
        p_j=p_i+year+"/"
        map_yr_list=os.listdir(p_j)
        for file in map_yr_list:
            p_k=p_j+file
            Data=open(p_k,'r')
            D=json.load(Data)
            for i in D['data']['data']['data']:
                lat = (i[0])
                column4['lat'].append(lat)
                long = (i[1])
                column4['lng'].append(long)
                metric = (i[2])
                column4['metric'].append(metric)
                District = (i[3])
                column4['label'].append(District)
                column4['state'].append(state)
                column4['year'].append(int(year))
                column4['quarter'].append(int(file.strip('.json')))
map_Insurance=pd.DataFrame(column4)
map_Insurance = map_Insurance.drop_duplicates(subset=['label'])
map_Insurance.reset_index(drop=True, inplace=True)
map_Insurance['label'] = map_Insurance['label'].str.title()

name_alter(map_Insurance)
map_Insurance['label'] = map_Insurance['label'].str.replace(' District', '')  
map_Insurance['new_label']= map_Insurance['state']+map_Insurance['label']
map_Insurance
```

```python
### map Transcation

path = "F:/phonepay/pulse/data/map/transaction/hover/country/india/state"
map_state_list=os.listdir(path)

column5 = {'state':[], 'year':[],'quarter':[], 'name': [], 'amount':[],'count':[]}
for state in map_state_list:
    p_i = path+'/'+ state +'/'
    map_yr = os.listdir(p_i)
    for year in map_yr:
        p_j=p_i+year+"/"
        map_yr_list=os.listdir(p_j)
        for file in map_yr_list:
            p_k=p_j+file
            Data=open(p_k,'r')
            D=json.load(Data)
            # print(D)
            for i in D['data']['hoverDataList']:
                name = (i['name'])
                count = (i['metric'][0]['count'])
                amount = (i['metric'][0]['amount'])
                column5['name'].append(name)
                column5['count'].append(count)
                column5['amount'].append(amount)
                column5['state'].append(state)
                column5['year'].append(int(year))
                column5['quarter'].append(int(file.strip('.json')))
map_trans=pd.DataFrame(column5)
map_trans['name'] = map_trans['name'].str.title()
name_alter(map_trans)
map_trans['name'] = map_trans['name'].str.replace(' District', '')  
map_trans['new_label']= map_trans['state']+map_trans['name']

map_trans
```

```python
## map user
path = "F:/phonepay/pulse/data/map/user/hover/country/india/state"
map_state_list=os.listdir(path)

column6 = {'state':[], 'year':[],'quarter':[], 'name': [], 'registeredusers':[],'appopens':[]}
for state in map_state_list:
    p_i = path+'/'+ state +'/'
    map_yr = os.listdir(p_i)
    for year in map_yr:
        p_j=p_i+year+"/"
        map_yr_list=os.listdir(p_j)
        for file in map_yr_list:
            p_k=p_j+file
            Data=open(p_k,'r')
            D=json.load(Data)
            # pprint(D)
            for i,j in D['data']['hoverData'].items():
                name = i
                registeredUsers = (j['registeredUsers'])
                appOpens = (j['appOpens'])

                column6['name'].append(name)
                column6['registeredusers'].append(registeredUsers)
                column6['appopens'].append(appOpens)
                column6['state'].append(state)
                column6['year'].append(int(year))
                column6['quarter'].append(int(file.strip('.json')))
map_user=pd.DataFrame(column6)
map_user['name'] = map_user['name'].str.title()
name_alter(map_user) 
map_user['name'] = map_user['name'].str.replace(' District', '') 
map_user['new_label']= map_user['state']+map_user['name']         
map_user
```
